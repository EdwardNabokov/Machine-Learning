```python
import numpy as np
import os
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from keras.datasets import boston_housing
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout, Activation
from keras.layers import MaxPooling2D, Convolution2D
from keras.utils import np_utils
from keras.optimizers import SGD
```

# Data

```python
(X_train, y_train), (X_test, y_test) = boston_housing.load_data()
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

# Model Keras

```python
model = Sequential()
```

```python
model.add(Dense(14, input_dim=13, kernel_initializer='normal', activation='relu'))
```

```python
model.add(Dense(8, activation='relu', use_bias=True))
```

```python
model.add(Dense(8, activation='relu', use_bias=True))
```

```python
model.add(Dense(1))
```

```python
sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=200)
```

```python
predicted = model.predict(X_test)
```

```python
print('Predicted (first 5): {}'.format(predicted.ravel()[:5]))
print('Actual (first 5): {}'.format(y_test[:5]))
```
